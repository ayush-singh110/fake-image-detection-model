```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y1=[45,65,85,96,74,25]
y2=[15,25,3,74,56,41]
plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color="red")
plt.title("Plot1")

plt.subplot(2,2,2)
plt.plot(y1,y2,color="green")
plt.title("Plot2")

z=[k**2 for k in y1]

plt.subplot(2,2,3)
plt.plot(z,y1,color="blue")
plt.title("Plot 3")

m=[v**2 for v in y1]
plt.subplot(2,2,4)
plt.plot(m,y2,color="grey")
plt.title("Plot 4")
```
